```python
import pandas as pd
import glob
import os
import pathlib
import logging
import urllib.parse as parse
```

```python
path = "all_csv_data//"
all_files = glob.glob(os.path.join(path, "*.csv"))

all_df = []
for f in all_files:
    # try:
    #     df = pd.read_csv(f, sep=',')
    # except:
    #     print(f)
    #     df = pd.read_csv(f, sep=',', error_bad_lines=False, warn_bad_lines=True)
    # 19 and 20 have some wrong rows
    df = pd.read_csv(f, sep=',', on_bad_lines="skip")
    df['file'] = f.split('\\')[-1]
    all_df.append(df)
    
merged_df = pd.concat(all_df, ignore_index=True, sort=False)
```

```python
for i in set(merged_df['AwayTeam'].values):
    club = str(i).split()[-1]
    club = parse.quote(club)
    path = f'http://api.clubelo.com/{club}'
    print(path)
    try:
        df = pd.read_csv(path)
    except:
        logging.warning(f"Invalid URL for club {i}")
    if df.empty:
        logging.warning(f"Can't find elo for club {i}")
    else:
        df.to_csv(f"club_elo_historical//{i}.csv")
```
