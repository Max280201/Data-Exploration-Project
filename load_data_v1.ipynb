```python
def elogain_2(z, hometeam, awayteam):
    homeelo = teams_elo_avstaerke[hometeam][0]
    awayelo = teams_elo_avstaerke[awayteam][0]

    trueoutcome = outcometrue_2(z)
    expectedoutcome = outcomeexpected_2(z, homeelo, awayelo)
    ofvictorymargin = marginofvictory_2(z, homeelo, awayelo)

    elogain = 20 * (trueoutcome - expectedoutcome) * ofvictorymargin
    elohome = homeelo + elogain
    eloaway = awayelo - elogain

    teams_elo_avstaerke[hometeam][0] = elohome
    teams_elo_avstaerke[awayteam][1] = eloaway

    return elohome, eloaway, homeelo, awayelo


def marginofvictory_2(z, homeelo, awayelo):
    elodiff = diffelo(homeelo, awayelo)
    diffgoal = data_1.FTHG[z] - data_1.FTAG[z]

    if diffgoal <= 1:
        marginofvictory = 1
    else:
        marginofvictory = (math.log2(1.7 * diffgoal) * 2) / (2 +
                                                             0.001 * elodiff)
    return marginofvictory


def outcometrue_2(z):
    if data_1.FTR[z] == "H":
        outcometrue = 1
    elif data_1.FTR[z] == "D":
        outcometrue = 0.5
    else:
        outcometrue = 0
    return outcometrue


def outcomeexpected_2(z, homeelo, awayelo):
    elodiff = diffelo(homeelo, awayelo)
    outcomeexpected = 1 / (1 + 10**(-elodiff / 400))
    return outcomeexpected
```

```python
def home_attack(z, alpha, ratio, hometeam, awayteam):
    goalsshot = data_2.FTHG[z]
    attack_old = teams_elo_avstaerke[hometeam][2]
    otherteamdefend = teams_elo_avstaerke[awayteam][3]
    homeattack = attack_old + ((goalsshot - attack_old) * ratio + (goalsshot + otherteamdefend) * (1-ratio)) * alpha
    teams_elo_avstaerke[hometeam][2] = homeattack
    return homeattack, attack_old

def home_defend(z, alpha, ratio, hometeam, awayteam):
    goalsgot = data_2.FTAG[z]
    defend_old = teams_elo_avstaerke[hometeam][3]
    otherteamattack = teams_elo_avstaerke[awayteam][2]
    homedefend = defend_old - ((goalsgot + defend_old) * ratio + (goalsgot - otherteamattack) * (1-ratio)) * alpha
    teams_elo_avstaerke[hometeam][3] = homedefend
    return homedefend, defend_old
```

## How good are bookmarkes?



```python
data_1["B365Prediction"] = data_1.apply(lambda x: 'H' if x.B365H < x.B365A and x.B365H < x.B365D else 'A' if x.B365A < x.B365D else 'D', axis=1)
data_1['BetTrue'] = data_1.B365Prediction.eq(data_1.FTR)
data_1.dropna(subset=['BetTrue', 'B365Prediction', 'FTR'])
accuracy_bet365 = data_1.BetTrue.value_counts()
data_1[['HomeTeam', 'AwayTeam', 'Date', 'B365Prediction', 'FTR', 'B365H', 'B365A', 'B365D']].head(n=20)
```

```python
print(f"Acc {round(accuracy_bet365[1]/(accuracy_bet365[0]+accuracy_bet365[1]), 10)}")
accuracy_bet365
```
